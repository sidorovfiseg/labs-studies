## Описание датасета
- __country__ - страна, где измерялась погода
- __location_name__ - название местоположения (город)
- __latitude__ - координаты широты
- __longitude__ - координаты долготы
- __timezone__ - часовой пояс
- __last_updated_epoch__ - временная метка последнего обновления данных в unix
- __last_updated__ - местное время последнего замера
- __temperature_celsius__ - температура в Цельсиях
- __temperature_fahrenheit__ - температура в Фаренгейтах 
- __condition_text__ - описание погодных условий
- __wind_mph__ - скорость ветра в милях в час
- __wind_kph__ - скорость ветра в километрах в час
- __wind_degree__ - направления ветра в градусах
- __wind_direction__ - направление ветра по 16-точечному компасу
- __pressure_mb__ - давление в миллибарах
- __pressure_in__ - давление в дюймах
- __precip_mm__ - количество осадков в миллиметрах 
- __precip_in__ - количество осадков в дюймах 
- __humidity__ - влажность в процентах
- __cloud__ - облачный покров в процентах
- __feels_like_celsius__ - температура по ощущениям в Цельсиях
- __feels_like_fahrenheit__ - температура по ощушениям в Фаренгейтах 
- __visibility_km__ - видимость в километрах
- __visibility_miles__ - видимость в милях
- __uv_index__ - ультрафиолетовый индекс
- __gust_mph__  - порыв ветра в милях в час
- __gust_kph__ - порыв ветра в километрах в час
- __air_quality_Carbon_Monoxide__ - измерение качества воздуха (монооксид углерода)
- __air_quality_Ozone__ - измерение качества воздуха (озон)
- __air_quality_Nitrogen_dioxide__ - измерение качества воздуха (диоксид азота)
- __air_quality_Sulphur_dioxide__ - измерение качества воздуха (диоксид серы)
- __air_quality_PM2.5__ - измерение качества воздуха (дисперсные вещества меньше 2.5 микрон)
- __air_quality_PM10.0__ - измерение качества воздуха (дисперсные вещества меньше 10.0 микрон)
- __air_quality_us-epa-index__ - измерение качества воздуха (американский индекс качества воздуха)
- __air_quality_gb-defra-index__ - измерение качества воздуха (gb defra индекс)
- __sunrise__ - местное время рассвета
- __sunset__ - местное время заката
- __moonrise__ - местное время восхода луны
- __moonset__ - местное время захода луны
- __moon_phase__ - текущая фаза луны
- __moon_illumination__ - процент освещенности луной

[Ссылка на датасет](https://www.kaggle.com/datasets/nelgiriyewithana/global-weather-repository/data?select=GlobalWeatherRepository.csv)

## Запросы

1. Вывод минимальной температуры в России

```python
rus_weather = df.loc[df['country'] == "Russia", ["country", "temperature_celsius", "location_name", "last_updated"]]

rus_weather[rus_weather['temperature_celsius'] == rus_weather["temperature_celsius"].min()]
```

2. Вывод мест, где средний уровень загрезнения воздуха 

```python
df.loc[(df['air_quality_us-epa-index'] > 3) & (df['air_quality_us-epa-index'] < 7), 
       ['country', 'location_name', 'air_quality_us-epa-index', 'last_updated']]
```

3. Вывод мест, в которых не было восхода луны

```python
df.loc[(df['moonrise'] == "No moonrise"), ["country", "location_name", "last_updated"]]
```

4. Количество мест, в которых восход был раньше 4:55 утра

```python

early_sunrise = df[(df['sunrise'].dt.hour < 5) & (df['sunrise'].dt.minute < 55)]

print("Количество мест, в которых восход был раньше 4:55 утра:", len(early_sunrise.groupby("location_name").agg({'country': 'unique', 'sunrise': 'unique'})))
```

5. Вывод страны с максимальной средней температурой и минимальной средней температурой

```python
avg_temp_countries = df.groupby(by="country")['temperature_celsius'].mean()
print(avg_temp_countries, "\n")
print(f"Страна с максимальной средней температурой: \
      {avg_temp_countries.idxmax()} --> {avg_temp_countries.max()}")

print(f"Страна с максимальной средней температурой: \
      {avg_temp_countries.idxmin()} --> {avg_temp_countries.min()}")
```

6. Средняя температура в зависимости от индекса ультрафиолетового излучения

```python
group_by_uv = df.groupby(by="uv_index")

group_by_uv['temperature_celsius'].mean()
```

7. Вывести топ 10 мест с наибольшей влажностью

```python
df.nlargest(10, "humidity")[[
  'location_name',
  "country",
  'humidity'
]]
```
